```python
import os

import requests
```

# Prepare

```python
auth_url = "https://ngw.devices.sberbank.ru:9443/api/v2/oauth"
basic_key = 'MWNhZmZkYzEtODNhNy00OTQxLThiNDktMGQyYjFjY2JhYmJjOjg5OGQ5OGQyLTc3Y2ItNDllNS05NThiLTkzYzBkODliYjIzOA=='
transcribe_url = "https://smartspeech.sber.ru/rest/v1/speech:recognize"
verify = False
```

# File

```python
audio_sets_path = os.path.join(os.path.dirname(os.getcwd()), 'Phonetics Project (Praat)')
in_a_restaurant_path = os.path.join(audio_sets_path, '1_In a restaurant.mp3')
in_a_restaurant_path
```

```python
with open(in_a_restaurant_path, 'rb') as f:
    file_data = f.read()
```

# Auth

```python
auth_payload={
  'scope': 'SALUTE_SPEECH_PERS'
}
auth_headers = {
  'Content-Type': 'application/x-www-form-urlencoded',
  'Accept': 'application/json',
  'RqUID': '3cb18962-ff84-4761-b656-b6614364f4b3',
  'Authorization': f'Basic {basic_key}'
}

response = requests.post(
    auth_url, 
    headers=auth_headers, 
    data=auth_payload, 
    verify=verify
)

access_token = response.json()['access_token']
```

# Transcribe

```python
query_params = {
    'language': 'en-US',
}

headers = {
  'Content-Type': 'audio/mpeg',
  'Accept': 'application/json',
  'Authorization': f'Bearer {access_token}'
}

response = requests.post(
    transcribe_url, 
    headers=headers, 
    data=file_data, 
    params=query_params, 
    verify=verify
)

response.json()
```

```python
text = response.json()['result']
len(text)
```
