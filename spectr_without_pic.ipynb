```python
import os
from typing import NamedTuple

import IPython.display as ipd
import librosa
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
class AudioSet(NamedTuple):
    duration: float
    file_name: str
    y: np.ndarray
    sr: int
```

```python
audio_sets_path = os.path.join(os.path.dirname(os.getcwd()), 'Phonetics Project (Praat)')
audio_sets_path
```

```python
audio_sets = []
for file_name in os.listdir(audio_sets_path):
    if file_name.endswith(('wav', 'mp3')):
        file_path = os.path.join(audio_sets_path, file_name)
        y, sr = librosa.load(file_path)
        duration = librosa.get_duration(y=y, sr=sr)
        audio_set = AudioSet(
            duration=duration,
            file_name=file_name,
            y=y,
            sr=sr
        )
        audio_sets.append(audio_set)
```

```python
grouping_by_duration = {_: [] for _ in range(1, 7)}
for a_s in audio_sets:
    if a_s.duration <= 30:
        grouping_by_duration[1].append(a_s)
    elif a_s.duration <= 60:
        grouping_by_duration[2].append(a_s)
    elif a_s.duration <= 90:
        grouping_by_duration[3].append(a_s)
    elif a_s.duration <= 120:
        grouping_by_duration[4].append(a_s)
    elif a_s.duration <= 150:
        grouping_by_duration[5].append(a_s)
    else:
        grouping_by_duration[6].append(a_s)
```

## 0 < duration <= 30

```python
audio_subsets = grouping_by_duration[1]
fig, ax = plt.subplots(nrows=len(audio_subsets), sharex=True)
fig.set_size_inches(30 * 2, 5 * len(audio_subsets))
for subplot, a_s in zip(ax, audio_subsets):
    subplot.set_title(a_s.file_name)
    D = np.abs(librosa.stft(a_s.y))
    librosa.display.specshow(
        librosa.amplitude_to_db(D, ref=np.max),
        x_axis='time',
        y_axis='log',
        ax=subplot
    )
plt.savefig('1.png')
```
